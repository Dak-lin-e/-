```python
### 객체지향 / 인공지능 설계 프로젝트 HW1
### 학번 : 20XXXXXX
### 이름 : 000
### 제출일자 : 22-09-XX

import random

def board_visualization(board):
    ### 틱택토 게임 출력 함수  
    for i_idx in range(3):
        for j_idx in range(3):
            
            if board[i_idx][j_idx] == 0: # not occupied
                print("▨", end='')
            elif board[i_idx][j_idx] == 1: # user
                print("○", end='')
            elif board[i_idx][j_idx] == 2: # computer
                print("×", end='')
            else:
                pass
        print()        
    
def board_turn(board, row, col, val):
    
    if 0 <= row < 3 and 0 <= col < 3:
        if board[row][col] == 0:
            board[row][col] = val
            return 1
        else:
            print("그곳에 돌을 둘 수 없습니다. 다시 입력하세요.")
            return 0
    else:
        print("그곳에 돌을 둘 수 없습니다. 다시 입력하세요.")
        return 0
        
    
def win_condition(v1,v2,v3): # 사용자 승리 return 100, 컴퓨터 승리 return -100, 무승부 return 0
    if v1 == v2 == v3:  
        if v1 == 1: return 100
        elif v2 == 2: return -100
        else: return 0
    else:
        return 0
        
    
def game_result(board): # 사용자 승리 return 100, 컴퓨터 승리 return -100, 무승부 return 0
    result = 0
    # check rows
    for i_idx in range(3):
        result = win_condition(board[i_idx][0],board[i_idx][1],board[i_idx][2])      
        if result != 0: return result
      
    # check cols
    for i_idx in range(3):
        result = win_condition(board[0][i_idx],board[1][i_idx],board[2][i_idx])
        if result != 0: return result
        
    #check diagonals
    result = win_condition(board[0][0],board[1][1],board[2][2])      
    if result != 0: return result
    result = win_condition(board[0][2],board[1][1],board[2][0])      
    if result != 0: return result
        
    return result

def position():
    row = int(input("돌을 놓을 행을 입력하세요 (0~2):"))
    col = int(input("돌을 놓을 열을 입력하세요 (0~2):"))
    
    return row, col


def minimax(board, turn):
    # board : 현재 틱택토 보드
    # turn : user 또는 computer의 차례
    
    # hint
    # 1. minimax함수에서는 사람이 이기면 큰 수 (+100), 컴퓨터가 이기면 작은수 (-100), 비기면 중간값(0) 반환 - game_result() 함수 활용
    # 2. ① 사용자가 둔 수 이후에 대한 탐색과 ② 컴퓨터가 둔 수 이후에 대한 탐색을 재귀적’으로 수행
    #    단, 사용자는 값을 최대화 하는쪽으로 / 컴퓨터는 값을 최소화 하는 쪽으로 탐색 
    #    재귀적으로 탐색을 수행하다보면 승패가 결정된 값이 반환됨
    
    
    return minimax_val

    
def computer(board):
    
                
    def win(board):              # 내가 이길수 있는 상황에서는 이기고 그 상황이 아니면 상대방이 이기는 경우의수를계산 후 막는 함수                    
        nonlocal row, col       # 두개의 상황이 없을 시에는 랜덤으로 수를 놓는 함수       
        empty_cnt = 0                              
        for i in range(3):       #board의상황을 체크하는 함수
            for j in range(3):
                if board[i][j]==0:     
                    if i == 0 and j == 0: # board i,j=0에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음 
                        if board[0][1] == board[0][2] and ( board[0][1] == 2):
                            row = 0
                            col = 0
                            return row, col
                        elif board[1][0] == board[2][0] and (board[1][0] == 2):
                            row = 0
                            col = 0
                            return row, col
                        elif board[1][1] == board[2][2] and ( board[1][1] == 2):
                            row = 0
                            col = 0
                            return row, col
                    elif i == 1 and j == 0:  # board i=1,j=0에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음 
                        if board[0][0] == board[2][0] and ( board[0][0] == 2):
                            row = 1
                            col = 0
                            return row, col
                        elif board[1][1] == board[1][2] and ( board[1][1] == 2):
                            row = 1
                            col = 0
                            return row, col
                    elif i == 2 and j == 0: # board i=2,j=0에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음
                        if board[0][0] == board[1][0] and ( board[0][0] == 2):
                            row = 2
                            col = 0
                            return row, col
                        elif board[2][1] == board[2][2] and ( board[2][1] == 2):
                            row = 2
                            col = 0
                            return row, col
                        elif board[1][1] == board[0][2] and (board[1][1] == 2):
                            row = 2
                            col = 0
                            return row, col
                       
                    elif i == 0 and j == 1: # board i=0,j=1에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음
                        if board[0][0] == board[0][2] and (board[0][0] == 2 ):
                            row = 0
                            col = 1
                            return row, col
                        elif board[1][1] == board[2][1] and (board[1][1] == 2 ):
                            row = 0
                            col = 1
                            return row, col
                            
                    elif i == 2 and j == 1: # board i=2,j=1에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음
                        if board[2][0] == board[2][2] and (board[2][0] == 2):
                            row = 2
                            col = 1
                            return row, col
                        elif board[1][1] == board[0][1] and (board[1][1] == 2 ):
                            row = 2
                            col = 1  
                            return row, col
                    elif i == 0 and j == 2:# board i=0,j=2에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음
                        if board[0][0] == board[0][1] and (board[0][0] == 2):
                            row = 0
                            col = 2
                            return row, col
                        elif board[1][2] == board[2][2] and (board[1][2] == 2):
                            row = 0
                            col = 2
                            return row, col
                        elif board[1][1] == board[2][0] and (board[1][1] == 2):
                            row = 0
                            col = 2  
                            return row, col
                    elif i == 1 and j == 2:# board i=1,j=2에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음
                        if board[1][0] == board[1][1] and (board[1][0] == 2):
                            row = 1
                            col = 2
                            return row, col
                        elif board[0][2] == board[2][2] and (board[0][2] == 2):
                            row = 1
                            col = 2  
                            return row, col
                    elif i == 2 and j == 2:# board i=2,j=2에  0이 있다면 그 위치에서 내가 이길 수 있는 상황인지 아닌지 체크하고 이길수 있는 상황이면 그 자리에 수를 놓음
                        if board[2][0] == board[2][1] and (board[2][0] == 2):
                            row = 2
                            col = 2
                            return row, col
                        elif board[0][2] == board[1][2] and (board[0][2] == 2):
                            row = 2
                            col = 2
                            return row, col
                        elif board[1][1] == board[0][0] and (board[1][1] == 2 ):
                            row = 2
                            col = 2
                            return row, col
                    else:
                        pass
                        
                            
        for i in range(3):   # 만약 이길 수 있는 경우의수가 없다면 상대방이 이길수 있는 경우의수를 체크하고 그것을 막는 코드
            for j in range(3):
                if board[i][j]==0:    
                    if i == 0 and j == 0: # board i=0,j=0에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[0][1] == board[0][2] and ( board[0][1] == 1):
                            row = 0
                            col = 0
                            return row, col
                        elif board[1][0] == board[2][0] and (board[1][0] == 1):
                            row = 0
                            col = 0
                            return row, col
                        elif board[1][1] == board[2][2] and ( board[1][1] == 1):
                            row = 0
                            col = 0
                            return row, col
                    elif i == 1 and j == 0:# board i=1,j=0에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[0][0] == board[2][0] and ( board[0][0] == 1):
                            row = 1
                            col = 0
                            return row, col
                        elif board[1][1] == board[1][2] and ( board[1][1] == 1):
                            row = 1
                            col = 0
                            return row, col
                    elif i == 2 and j == 0:# board i=2,j=0에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[0][0] == board[1][0] and ( board[0][0] == 1):
                            row = 2
                            col = 0
                            return row, col
                        elif board[2][1] == board[2][2] and ( board[2][1] == 1):
                            row = 2
                            col = 0
                            return row, col
                        elif board[1][1] == board[0][2] and (board[1][1] == 1):
                            row = 2
                            col = 0
                            return row, col
                       
                    elif i == 0 and j == 1:# board i=0,j=1에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[0][0] == board[0][2] and (board[0][0] == 1 ):
                            row = 0
                            col = 1
                            return row, col
                        elif board[1][1] == board[2][1] and (board[1][1] == 1 ):
                            row = 0
                            col = 1
                            return row, col
                            
                    elif i == 2 and j == 1:# board i=2,j=1에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[2][0] == board[2][2] and (board[2][0] == 1):
                            row = 2
                            col = 1
                            return row, col
                        elif board[1][1] == board[0][1] and (board[1][1] == 1 ):
                            row = 2
                            col = 1  
                            return row, col
                    elif i == 0 and j == 2:# board i=0,j=2에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[0][0] == board[0][1] and (board[0][0] == 1):
                            row = 0
                            col = 2
                            return row, col
                        elif board[1][2] == board[2][2] and (board[1][2] == 1):
                            row = 0
                            col = 2
                            return row, col
                        elif board[1][1] == board[2][0] and (board[1][1] == 1):
                            row = 0
                            col = 2  
                            return row, col
                    elif i == 1 and j == 2:# board i=1,j=2에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[1][0] == board[1][1] and (board[1][0] == 1):
                            row = 1
                            col = 2
                            return row, col
                        elif board[0][2] == board[2][2] and (board[0][2] == 1):
                            row = 1
                            col = 2  
                            return row, col
                    elif i == 2 and j == 2:# board i=2,j=2에  0이 있다면 그 위치에서 상대방이 이길 수 있는 상황인지 아닌지 체크하고 상대방이 이길수 있는 상황이면 그 자리에 수를 놓아서 상대방이 이기는 걸 막음
                        if board[2][0] == board[2][1] and (board[2][0] == 1):
                            row = 2
                            col = 2
                            return row, col
                        elif board[0][2] == board[1][2] and (board[0][2] == 1):
                            row = 2
                            col = 2
                            return row, col
                        elif board[1][1] == board[0][0] and (board[1][1] == 1 ):
                            row = 2
                            col = 2
                            return row, col
                        else: 
                            pass
                else:                 # 위의 2개의 상황이 아닐시 랜덤으로 아무 곳에나 수를 놓는 코드

                    for i_idx in range(3):
                        for j_idx in range(3):
                            if board[i_idx][j_idx] == 0:
                                empty_cnt+=1

                    rnd_idx = random.randrange(1,empty_cnt)

                    empty_cnt = 0

                    for i_idx in range(3): 
                        for j_idx in range(3):
                            if board[i_idx][j_idx] == 0:          
                                if empty_cnt == rnd_idx:
                                    row = i_idx
                                    col = j_idx
                                    return row, col
                                empty_cnt+=1


                    #return row, col

                    
                            
                            
                        
        
        

                
    empty_n = 0                         #board 안에 놓여진 수를 계산하여 상황에맞는 경우의수를 적용하기 위해 만든 변수
    row = 0
    col= 0
    for i in range(3):                  #board 안에 놓여진 수를 계산하는 for 문
        for j in range(3):
            if board[i][j] == 0:
                empty_n+=1
                
                
                
    if empty_n==8 : # 상대방이 첫 수를 놓은 상황
        for i in range(3):
            for j in range(3):
                if board[i][j] == 1: #상대방이 중앙에 첫수를 놓은 상황이므로 대각선라인 아무곳에나 수를놓으면 되므로 row=2 col=0에 수를놓음
                    if i==1 and j==1 : #만약 대각선 라인에 놓지 않을시 질 수 밖에 없는 상황이 만들어진다. 자세한건 첨부파일에서 설명
                        row=2 
                        col= 0
                        return row, col
                    elif i!=1 or j!=1: #만약 상대방이 중앙에 첫수를 놓지 않았다면 row=1 col=1을 줘서 중앙에 수를 놓음
                        row=1           #중앙에 수를 놔둬야 대각선으로 상대방이 이기는 걸 원천 차단할 수 있음. 한 마디로 상당히 유리해짐
                        col=1
                        return row, col
    elif empty_n==6 :                   # 4번쨰 수를 둘 차례가 옴.(나에겐 두번 쨰 차례)
        if board[2][0]==2:              # 내가 중앙에 수를 뒀는지 안 뒀는지 확인하기 위함 이전에 row=2 col=0으로 설정을 해서 이 코드가 맞으면 상대방이 중앙을 먹은 상황
             for i in range(3):
                for j in range(3):
                    if board[i][j] == 1: # 3번째수에서 상대방이 어디다 뒀는지 체크하기 위한 코드 
                        if i==0 and j==0 : # 만약 중앙을 먹힌 상황에서 3번쨰수로 상대방이 i,j=0인 곳에 수를 놨으면  
                            row=2           # 반드시 row=2 col=2에다가 수를 놔야됨. 그렇지 않으면 이후에 무조건 지는 상황이 만들어질 수 있음 
                            col=2
                            return row, col
                        elif i==0 and j==2:# 만약 중앙을 먹힌 상황에서 3번쨰수로 상대방이 i=0,j=2인 곳에 수를 놨으면
                            row=2            # 반드시 row=2 col=2에다가 수를 놔야됨. 그렇지 않으면 이후에 무조건 지는 상황이 만들어질 수 있음
                            col=2
                            return row, col
                        elif i==2 and j==2:# 만약 중앙을 먹힌 상황에서 3번쨰수로 상대방이 i=2,j=2인 곳에 수를 놨으면
                            row=0           # 반드시 row=0 col=0에다가 수를 놔야됨. 그렇지 않으면 이후에 무조건 지는 상황이 만들어질 수 있음
                            col=0
                            return row, col
                        elif i%2==0 and j==1: # 만약 중앙을 먹힌 상황에서 3번쨰수로 상대방이 십자가라인에 수를 놨으면
                            win(board)          # 다행히 이후에는 win 함수에 맞겨서 흘러가는대로 수를 놔둬도 이기거나 비길 수 있음
                            return row, col
                        elif i==1 and j%2==0:
                            win(board)
                            return row, col
                       
                            
                        else :
                            pass
                             
                           
                        
        elif board[1][1]==2:      # 위에서 말했듯이 내가 중앙을 먹으면 상당히 유리해짐. 허나, 한가지 경우의수를 제외해야함. 
            if board[0][0]==board[2][2] and (board[0][0]==1): # 중앙을 먹어도 옆에 있는 상황에서 대각선 라인에 4번쨰수 를 놓게되면
                row=1                                          # 무조건 지는 상황이 만들어 질 수 있음. 따라서 십자가라인에 수를 놔둬야함.
                col=0                                          #그래서 row=1 col=0에 수를 놔두는 것으로 설정함.
                return row,col
            elif board[0][2]==board[2][0] and (board[0][2]==1): # 이 옆에 있는 상황에서도 4번쨰수를 십자가라인에 놔야함.
                row=1                                            # 안그러면 이후에 무조건 지는 상황이 만들어질 수 있음.
                col=0
                return row,col
            else:
                win(board)           # 위의 상황이 아니라면 win 함수를 사용해서 흘러가는 대로 두면 이기거나 비길 수 있음.
            
                   
    elif 0 < empty_n < 6:  # 위의 무조건 지는 상황들을 피하면 win 함수에 맞겨서 흘러가는대로 두면 이기거나 비길 수 있음.
            win(board)
             
    else:
        pass
        

    
    return row, col       
    
   


   
    
        

### 게임 시작
print("게임 시작")
board = [[0,0,0],[0,0,0],[0,0,0]]
board_visualization(board)

for i_idx in range(5): # 플레이어는 최대 5번 돌을 둔다
    ###### User ######
    print("============= Your turn =============")
    row, col = position()
    while board_turn(board, row, col,1) == 0:
        row, col = position()
        
    board_visualization(board)
    if game_result(board) == 100:
        print("사용자가 승리하였습니다.")
        break
    elif game_result(board) == -100:
        print("컴퓨터가 승리하였습니다.")
        break      
    ########################
    
    if i_idx == 4: #9번을 넘어가서 두는 경우 제외
        break
    
    ###### Computer ######
    print("============= Computer =============")
    row, col = computer(board) # computer 함수를 다시 코딩 하시오
    board_turn(board, row, col, 2)
    board_visualization(board)   
    if game_result(board) == 100:
        print("사용자가 승리하였습니다.")
        break
    elif game_result(board) == -100:
        print("컴퓨터가 승리하였습니다.")
        break
    ########################

    
if game_result(board) == 0:
    print("비겼습니다.")
```
