```python
# 코드처리하기

# 문자열 code가 주어집니다.
#code를 앞에서부터 읽으면서 만약 문자가 "1"이면 mode를 바꿉니다. mode에 따라 code를 읽어가면서 문자열 ret을 만들어냅니다.
#mode는 0과 1이 있으며, idx를 0 부터 code의 길이 - 1 까지 1씩 키워나가면서 code[idx]의 값에 따라 다음과 같이 행동합니다.
# mode가 0일 때
# code[idx]가 "1"이 아니면 idx가 짝수일 때만 ret의 맨 뒤에 code[idx]를 추가합니다.
# code[idx]가 "1"이면 mode를 0에서 1로 바꿉니다.
# mode가 1일 때
# code[idx]가 "1"이 아니면 idx가 홀수일 때만 ret의 맨 뒤에 code[idx]를 추가합니다.
# code[idx]가 "1"이면 mode를 1에서 0으로 바꿉니다.
# 문자열 code를 통해 만들어진 문자열 ret를 return 하는 solution 함수를 완성해 주세요.

# 단, 시작할 때 mode는 0이며, return 하려는 ret가 만약 빈 문자열이라면 대신 "EMPTY"를 return 합니다.


def solution(code):
  ret=[]
  mode = False
  for i in code:
    if code.index(i) % 2 ==0 and mode == False:
      ret.append(i)
    elif i== '1' :
      mode= not mode
    elif code.index(i) % 2 !=0 and i!=1 and mode == True :
      ret.append(i)
    else :
      pass

  if len(ret)==0:
    ret.append('EMPTY')
  else:
    pass


  return ''.join(ret)

# 해결실패

# 실패한 이유:   .index(i) 는 처음 등장한 인덱스만 반환하기 때문. 즉, a 가 2번 나왔을때 2번쨰 a 도 처음나온 a의 인덱스로 반환하게됨.

# 지피티 답변
def solution(code):
    mode = 0
    ret = []

    for idx, char in enumerate(code):
        if char == '1':
            mode = 1 - mode  # mode 전환
        else:
            if mode == 0 and idx % 2 == 0:
                ret.append(char)
            elif mode == 1 and idx % 2 == 1:
                ret.append(char)

    return ''.join(ret) if ret else "EMPTY"


# 다른 사람풀이
def solution(code):
    return "".join(code.split("1"))[::2] or "EMPTY"
```

```python
# 배열 만들기

# 정수 l과 r이 주어졌을 때, l 이상 r이하의 정수 중에서 숫자 "0"과 "5"로만 이루어진 모든 정수를 오름차순으로 저장한 배열을 return 하는 solution 함수를 완성해 주세요.
# 만약 그러한 정수가 없다면, -1이 담긴 배열을 return 합니다

def solution(l, r):
  result=[]
  for i in range(l,r):        # l과 r 범위의 정수를 모두 집어넣고 숫자에 0이나 5가 없으면 탈락시킴.
    result.append(i)
    for j in str(i):
      if j != '0' and j!='5' :
        result.remove(i)
        break                 # 안쪽 for문만 탈출하는 break
      else:
        pass

  if len(result)==0:          # 정수가 없을시 -1 return
    result.append(-1)
  else:
    pass


  return result

# 다른 사람 풀이

def solution(l, r):
    answer = []
    for num in range(l, r + 1):
        if not set(str(num)) - set(['0', '5']):
            answer.append(num)
    return answer if answer else [-1]
```

```python
# 주사위게임3 - 조건문

# 1부터 6까지 숫자가 적힌 주사위가 네 개 있습니다. 네 주사위를 굴렸을 때 나온 숫자에 따라 다음과 같은 점수를 얻습니다.
# 네 주사위에서 나온 숫자가 모두 p로 같다면 1111 × p점을 얻습니다.
# 세 주사위에서 나온 숫자가 p로 같고 나머지 다른 주사위에서 나온 숫자가 q(p ≠ q)라면 (10 × p + q)2 점을 얻습니다.
# 주사위가 두 개씩 같은 값이 나오고, 나온 숫자를 각각 p, q(p ≠ q)라고 한다면 (p + q) × |p - q|점을 얻습니다.
# 어느 두 주사위에서 나온 숫자가 p로 같고 나머지 두 주사위에서 나온 숫자가 각각 p와 다른 q, r(q ≠ r)이라면 q × r점을 얻습니다.
# 네 주사위에 적힌 숫자가 모두 다르다면 나온 숫자 중 가장 작은 숫자 만큼의 점수를 얻습니다.
# 네 주사위를 굴렸을 때 나온 숫자가 정수 매개변수 a, b, c, d로 주어질 때, 얻는 점수를 return 하는 solution 함수를 작성해 주세요.


def solution(a, b, c, d):
    m=[a,b,c,d]
    n=[]        # 중복안된 값들만 저장할 리스트
    p=[]        # 중복된 값을 저장할 리스트
    for i in m:
      if i not in n:
          n.append(i)
      else :
          p.append(i)

    if len(n) == 1:
        answer = 1111 * a
    elif len(n) == 4:
        answer = min(n)
    elif len(n) == 3:
        if p[0]==n[0]:
            answer = n[1] * n[2]
        elif p[0]==n[1]:
            answer = n[0] * n[2]
        elif  p[0]==n[2] :
            answer = n[0] * n[1]
        else :
          pass

    elif len(p) == 2:
        if p[0] == p[1]:
            if p[0] !=n[0]:
                answer = (10 * p[0]+n[0])**2
            elif p[0] ==n[0]:
                answer = (10 * p[0]+n[1])**2
            else:
                pass
        elif p[0] != p[1]:
            answer = (p[0]+p[1]) * abs(p[0]-p[1])
        else :
            pass
    else :
        pass



    return answer
```

```python
# 정수를 나선형으로 배치하기

# 양의 정수 n이 매개변수로 주어집니다. n × n 배열에 1부터 n2 까지 정수를 인덱스 [0][0]부터 시계방향 나선형으로 배치한 이차원 배열을 return 하는 solution 함수를 작성해 주세요.

def solution(n):
  answer=[[]]
  i=0
  j=0
  while(True):
    for k in range(1,n**2):
      # answer[i][j] = k
      if i==0:
        answer[i][j]=k
        j++

      elif len(answer[0])==n :
        i++
        answer[i][j]=k

      elif 0 < i < n and len(answer[i])==1:
        i++
        answer[i][j]= k

      elif i ==n:
        j--
        answer[i][j]=k

      elif len(answer[n])==n and i ==n :
        i--
        # if len(answer[i])==n:
        #   i++
        #   j++
        #   answer[i][j]=k
        # else :
        answer[i][j]=k

      elif 0 < i < n and len(answer[i])==n:
        i++
        if len(answer[i]==n):
          i--
          j--
          answer[i][j]= k

      elif 0 < i < n and  1< len(answer[i]) < n:
        i--
        answer[i][j]=k

      elif k > n**2 :
        break

      else:
        pass


    return answer


# 해결실패 - 무지성 for문 반복. 알고리즘 공부 부족한듯. 결론은 걍 모르겠음.

# 지피티 풀이

def solution(n):
    matrix = [[0] * n for _ in range(n)]
    num = 1
    left, right, top, bottom = 0, n - 1, 0, n - 1

    while num <= n * n:
        # 왼쪽 → 오른쪽
        for i in range(left, right + 1):
            matrix[top][i] = num
            num += 1
        top += 1

        # 위 → 아래
        for i in range(top, bottom + 1):
            matrix[i][right] = num
            num += 1
        right -= 1

        # 오른쪽 → 왼쪽
        for i in range(right, left - 1, -1):
            matrix[bottom][i] = num
            num += 1
        bottom -= 1

        # 아래 → 위
        for i in range(bottom, top - 1, -1):
            matrix[i][left] = num
            num += 1
        left += 1

    return matrix
```

```python
# 배열회전시키기

# 정수가 담긴 배열 numbers와 문자열 direction가 매개변수로 주어집니다.
# 배열 numbers의 원소를 direction방향으로 한 칸씩 회전시킨 배열을 return하도록 solution 함수를 완성해주세요.

def solution(numbers, direction):
    n= len(numbers)
    answer =[0]*n         #  순서 바꿔줄 리스트 생성
    if direction == 'right':
        for i in range(n):
            answer[i-n+1]=numbers[i]
    elif direction == 'left':
        for i in range(n):
            answer[i-1]=numbers[i]
    else:
        pass

    return answer


# 다른 사람 풀이

from collections import deque

def solution(numbers, direction):
    numbers = deque(numbers)
    if direction == 'right':
        numbers.rotate(1)
    else:
        numbers.rotate(-1)
    return list(numbers)
```
